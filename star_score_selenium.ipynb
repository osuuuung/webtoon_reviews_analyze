```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import time
```

```python
# Set up the Chrome web driver (Update the executable path with your chromedriver location)
driver = webdriver.Chrome()
driver.implicitly_wait(3)
```

```python
total_page_list = [25,15,26,12,23,19]
titleid_list = [597447,702608,570506,701535,641253,670143]
title_list = ['프리드로우','랜덤채팅의 그녀','최강전설 강해효','격기 3반','외모지상주의','킬베로스']
```

```python
def Collect_Score(titleid,total_page, title):

    episode_list = []
    score_list = []
    name_list = []

    for page_num in range(1, total_page + 1):
        base_url = 'https://comic.naver.com/webtoon/list?titleId=597447&page={}&sort=DESC'
        driver.get(base_url.format(page_num))

        episode_tags = driver.find_elements(By.XPATH,'//*[@id="content"]/div[3]/ul/li/a')
        score_tags = driver.find_elements(By.XPATH,'//*[@id="content"]/div[3]/ul/li/a/div[2]/div/span[1]/span')

        for episode_tag in episode_tags:
            link = episode_tag.get_attribute('href')
            episode_number = link.split('no=')[1]
            episode_list.append(episode_number)

        for score_tag in score_tags:
            score = score_tag.text.strip()
            score_list.append(score)
            name_list.append(title)
        time.sleep(1)
    data = {'name' : name_list,'episode' : episode_list, 'score' : score_list}
    df = pd.DataFrame(data)
    return df
```

```python
for i in range(0,len(title_list)):
    
    titleid = titleid_list[i]
    title = title_list[i]
    total_page = total_page_list[i]
    
    score_df = Collect_Score(titleid,total_page, title)
    
    score_df = score_df.sort_values(by = ['score'], ascending = False)
    print(score_df)
    print('------')
    
    score_number = score_df['name'].count()
    collect_number = int(score_number*0.2)
    
    high_rank_df = score_df.head(collect_number)
    print(high_rank_df)
    print('------')
    low_rank_df = score_df.tail(collect_number)
    print(low_rank_df)
    print('------')
    high_rank_df.to_excel('{}_상위.xlsx'.format(title), sheet_name='score')
    low_rank_df.to_excel('{}_하위.xlsx'.format(title), sheet_name='score')
    time.sleep(3)
```

```python
# Close the web driver
driver.quit()
```
